```python
from sympy import init_session
from sympy import div, QQ
from sympy.parsing.latex import parse_latex
from sympy.parsing.sympy_parser import parse_expr
#from sympy.functions.combinatorial.numbers import nC, nP, nT
from IPython.display import Markdown as md
from IPython.display import display
import pandas as pd

init_session()

a, b, c, d = symbols('a b c d', real = True)
from docparts import añadir_ejercicios, doc_ejer, escribir_preambulo, escribir_ejercicios, escribir_fin

df_ejercicios = pd.DataFrame()

fichero = 'ej_sistemas'
titulo = 'Ejercicios de sistemas'
curso = '2BC'

```

```python
def sist_param(sist, k=k) :
  sist_latex = r"$\left\{ \begin{matrix}"
  for e in sist:
    sist_latex += latex(e)+r" \\ "  
  sist_latex += r"\end{matrix}\right.$"
  enunciado_latex = r"Discutir y resolver el siguiente sistema con parámetro ${}$: \\ ${}$ ".format(k,sist_latex)
  enunciado = str(sist)
  solucion = r""


  
  pprint(sist)
  pprint(enunciado)
  A, b = linear_eq_to_matrix(sist,[x,y,z])
  AA = A.row_join(b)
  AAs = AA.LUdecomposition()[1].applyfunc(simplify)
  
  pprint("Discusión y resolución por Gauss:")
  
  solucion += r"\textbf{Discusión y resolución por Gauss:} Escalonando la matriz ampliada tenemos\\"
  solucion += r"$A^*= {} \thicksim {}$. \\  De los valores de la última fila podemos concluir:".format(latex(AA),latex(AAs)).replace('[','(').replace(']',')')
  solucion += r"\begin{itemize}"
  pprint(AA.LUdecomposition()[1].applyfunc(simplify))
  pprint("det(A)={}".format(A.det()))
  pprint(AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-2:])
  for i in (solve(AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-2])) :
    #pprint(i)
    if AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-1].subs(k,i) == 0 :
      pprint("Si k = {} -->  0z=0 --> S.C.I".format(i))
      solucion += r"\item Si $k = {} \to$ $${}$$ La última fila es $0z=0 \to $ S.C.I".format(i,latex(AAs.subs(k,i))).replace('[','(').replace(']',')')
      pprint([eq.subs(k,i) for eq in sist])
      sol = list(zip([x,y,z],linsolve([eq.subs(k,i) for eq in sist],[x,y,z]).args[0],[r.subs(k,i) for r in [AA.LUdecomposition()[1].applyfunc(simplify).row(j) for j in range(AA.shape[0])]]))
      pprint(sol)
      for s in reversed(sol) :
        pprint("{} --> {} = {}".format(s[2],s[0],s[1].subs(z,"\lambda")))
        solucion += r"\begin{itemize}"
        solucion += r"\item ${} \to {} = {}$".format(latex(s[2]),latex(s[0]),latex(s[1]).replace('z',"\lambda")).replace('[','(').replace(']',')')
        solucion += r"\end{itemize}"
    else :
      pprint("Si k = {} --> 0z={} -->S.I".format(i, AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-1].subs(k,i)))
      solucion += r"\item Si $k = {} \to$ $${}$$ La última fila es $0z={} \to $ S.I.".format(i,latex(AAs.subs(k,i)),AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-1].subs(k,i)).replace('[','(').replace(']',')')
 
  pprint("si k<> {}  --> S.C.D.".format(solve(AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-2])))
  solucion += r"\item si $k\neq {}  \to $ S.C.D.".format(solve(AA.LUdecomposition()[1].applyfunc(simplify).row(-1)[-2]))

  pprint(list(linsolve(sist,[x,y,z]).args[0].args))
  sol = list(zip([x,y,z],linsolve(sist,[x,y,z]).args[0],[AA.LUdecomposition()[1].applyfunc(simplify).row(j) for j in range(AA.shape[0])]))
  pprint(sol)

  for s in reversed(sol) :
    pprint("{} --> {} = {}".format(s[2],s[0],s[1]))
    solucion += r"\begin{itemize}"
    solucion += r"\item ${} \to {} = {}$".format(latex(s[2]),s[0],latex(s[1])).replace('[','(').replace(']',')')
    solucion += r"\end{itemize}"
  
  solucion += r"\end{itemize}  "
  pprint("Por rangos y determinantes:")
  solucion += r"\textbf{Por rangos y determinantes:} \\"


  #Rango de A:
  solucion += r"$\left|A\right|={} \to \left|A\right|=0 \quad si \quad k = {}$".format(latex(A.det()),latex(solve(A.det())))
  solucion += r"\begin{itemize}"
  for i in solve(A.det()):

    A2 = A.subs(k,i)
    AA2 = AA.subs(k,i)
    pprint(A2)
    pprint(A2.rank())
    pprint(AA2.rank())
    if A2.rank() < AA2.rank() :
      pprint("Si k = {} --> S.I.".format(i))
      solucion += r"\item Si $k={} \to rg(A)={} \land rg(A^*)={} \to $ S.I.".format(i,A2.rank(),AA2.rank())
    else :
      pprint("Si k = {} --> S.C.I --> solo se puede resolver por Gauss, ver más arriba".format(i))
      solucion += r"\item Si $k={} \to rg(A)={} \land rg(A^*)={} \to $ S.C.I. $\to$ solo se puede resolver por Gauss, (ver más arriba)".format(i,A2.rank(),AA2.rank())

  pprint("Si k <> {} --> Rango(A)=Rango(A)={} --> S.C.D.".format(solve(A.det()),A.rank()))
  solucion += r"\item Si $k\neq{} \to rg(A)={} \land rg(A^*)={} \to $ S.C.D. $\to$ \\ ".format(solve(A.det()),A.rank(),AA.rank())
  pprint("Por Cramer:")
  sol=[]
  solucion += r"Por Cramer: \begin{itemize}"
  for i, var in enumerate([x,y,z]):
    #print("columna"+latex(i))
    AA.col_swap(i,3)
    pprint("Delta_i, A.det, solucion_i : ")
    sol.append([var,AA[:,:-1],simplify(AA[:,:-1].det()),simplify(A.det()),simplify(AA[:,:-1].det()/A.det())])
    pprint(sol[i])
    #print(r"$x_"+latex(i)+r"=\frac{"+latex(AA[:,:-1])+r"}{"+latex(A.det())+r"}=\frac{"+latex(AA[:,:-1].det())+r"}{"+latex(A.det())+r"}="+latex(AA[:,:-1].det()/A.det())+"$")
    solucion += r"\item $"+latex(sol[i][0])+r"=\frac{"+latex(sol[i][1]).replace('[','|').replace(']','|')+r"}{"+latex(sol[i][3])+r"}=\frac{"+latex(sol[i][2])+r"}{"+latex(sol[i][3])+r"}="+latex(sol[i][4])+r"$"
    AA.col_swap(i,3)

  solucion += r"\end{itemize}"
  solucion += r"\end{itemize}"
  
  # pprint("Enunciado latex:")
  # print(enunciado_latex)
  # pprint("Enunciado sympy:")
  # print(enunciado)
  # pprint("Solución:")
  # print(solucion)

  return(enunciado_latex,enunciado,solucion)


#[sist_param(i) for i in sistemas1]
#[sist_param(i,m) for i in sistemas2]
```

```python
# Página 12 ejercicios 3


sistemas = [
            [Eq(-3*x+2*y+3*z,-2),Eq(2*x+k*y-5*z,-4),Eq(x+y+2*z,2)],
           [Eq(k*x + y + z , k),Eq(x + k*y + z , k), Eq(x + y + k*z , k)],
            [Eq(x + y + z , k + 2),Eq(x - k*y + z , 1),Eq(k*x + y + z , 4)],
            [Eq(k*x+(k**2+1)*y+k*z,k),Eq(x+k*y+z,0),Eq(x+(k+1)*y+k**2*z,2*k-1)],
            [Eq(2*x - 5*y + 3*z , 0),Eq(x - y + z , 0),Eq(3*x + k*y + z , 0)],
            [Eq(x + k*y - z , 0),Eq(12*x - 3*y - 2*z , 0),Eq(x - 2*y + z , 0)],
            [Eq(x + y + 2*z , 0),Eq(k*x + y - z , k - 2), Eq(3*x + k*y + z , k - 2)],
            #[Eq(k*x+y,1),Eq(2*x+(k+1)*y,2),Eq(x+k*y,1)] son linealmente dependientes las filas de A ,
            ]

texto_ejercicio = r'Ejercicios de sistemas con un parámetro:'

n_ejercicio = ''

display(md(""+texto_ejercicio+""))


enunciado_latex, enunciado, solucion = [],[],[]
for s in sistemas :
    enl, en , sol = sist_param(s)
    enunciado_latex.append(enl)
    enunciado.append(en)
    solucion.append(sol)
    
df_ejercicios = df_ejercicios.append(añadir_ejercicios(enunciado_latex, enunciado, solucion, texto_ejercicio, curso, titulo, n_ejercicio, n_columnas=1, enun_tex=False, sol_tex=False), ignore_index=True)
```
