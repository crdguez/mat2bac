```python
from sympy import *
init_session()
from IPython.display import display_latex
```

```python
%matplotlib inline

datos = [-x**2+6*x-5, x*(x-1)**2, 2*x**3-15*x**2+36*x-12, 2/(1+x**2),(8*x)/(x**2+2),
         x*log(x), x**2*E**x, x/log(x), x-sin(x), x+cos(x), 2*sin(x)- sin(2*x)
        ]

for f in datos :
    display('----------------------------------------------------------------------------------------------------')
#     display([f,f.diff().simplify(),f.diff().diff().simplify()], [[i,f.diff().diff().subs(x,i)] for i in solve(f.diff())])
    display_latex("$f(x) =" + latex(f) + "$",raw=True)
    display_latex("$f'(x)=" + latex(f.diff().simplify()) + "$",raw=True)
    display_latex("$f''(x)=" + latex(f.diff().diff().simplify()) + "$",raw=True)
    for i in solve(f.diff()) : 
        display_latex("$f'("+latex(i)+")=0 \land f''("+latex(i)+")="+latex(f.diff().diff().subs(x,i))+"$",raw=True)
#         tex= "Hay un mínimo relativo" if f.diff().diff().subs(x,i) < 0   else "Hay un máximo relativo" if f.diff().diff().subs(x,i) < 0
        tex = "Hay un mínimo relativo" if f.diff().diff().subs(x,i) > 0 else "Hay un máximo relativo" if f.diff().diff().subs(x,i) < 0else  "No hay mínimo ni máximo"  
        display_latex(tex + ' en $('+ latex(i) +","+latex(f.subs(x,i))+')$', raw=True)
    plot(f)   
    
```
